```python
def NotaSkate():  # Fun√ß√£o que inicia o programa de notas da competi√ß√£o
    while True:  # Tem a fun√ß√£o de loop de escolha de op√ß√£o, at√© o usuario optar por sair
        print("\n          ===== <<< MODALIDADE >>>  =====")
        print("         |         \033[31m[1]\033[m Street            |")  # Op√ß√£o de modalidade exibidas
        print("         |         \033[31m[2]\033[m Park              |")
        print("         |         \033[31m[0]\033[m Sair              |")  # Op√ß√£o do usuario sair do programa
        print("         --------------------------------")
        Opcao = input("Insira a op√ß√£o: ")  # Solicita ao usuario inserir uma op√ß√£o

        if Opcao == "1":  # Se a op√ß√£o for 1, escolheu Street
            print("\nVoc√™ escolheu a op√ß√£o: Street\n")
            Sexo()  # Chama a fun√ß√£o Sexo para coletar o sexo do atleta
        elif Opcao == "2":  # Se a op√ß√£o for 2, escolheu Park
            print("\nVoc√™ escolheu a op√ß√£o: Park\n")
            Sexo()  # Chama a fun√ß√£o Sexo para coletar o sexo do atleta
        elif Opcao == "0":  # Se a op√ß√£o for 0, o usuario deseja sair
            print("Saindo do sistema...")
            break  # Sai do loop e finaliza o programa
        else:  # Se a op√ß√£o for inv√°lida ser√° mostrado a mensagem abaixo
            print("\nOp√ß√£o inv√°lida! Tente novamente.\n")  # Mostra ao usuario que a op√ß√£o escolhida fui inv√°lida, e solicita nova entrada de dados

def Sexo():  # Fun√ß√£o para escolher o sexo dos atletas
    while True:  # Inicia um loop para coletar o sexo
        print("\n          ===== <<< Sexo >>> =====")
        print("         |    \033[31m[1]\033[m Feminino        |")  # Op√ß√£o de sexo do atleta exibidas
        print("         |    \033[31m[2]\033[m Masculino       |")
        print("         |    \033[31m[0]\033[m Voltar          |")  # Op√ß√£o do usuario sair do programa
        print("         --------------------------")
        Opcao = input("Insira a op√ß√£o: ")  # Solicita ao usuario inserir uma op√ß√£o

        if Opcao in ["1", "2"]:  # Se a op√ß√£o for 1 ou 2, o usuario escolheu o sexo do atleta
            if Opcao == "1":  # Se a op√ß√£o for 1, escolheu o sexo Feminino
                print("\nVoc√™ escolheu a op√ß√£o: Feminino\n") # Informa a op√ß√£o escolhida
            else:  # Se a op√ß√£o for 2, escolheu o sexo Masculino
                print("\nVoc√™ escolheu a op√ß√£o: Masculino\n")  # Informa a op√ß√£o escolhida
            print("\n===== <<<.............................. >>> =====")
            NumAtletas = int(input("Quantos atletas participar√£o? "))  # Solicita ao usuario a quantidade de atletas participantes da competi√ß√£o
            print("\n===== <<<.............................. >>> =====")
            NumJuizes = 5  # Define que sempre ser√£o 5 ju√≠zes
            Juizes = ColetarNomesJuizes(NumJuizes)  # Coleta os nomes dos ju√≠zes
            ColetarNotas(NumAtletas, Juizes)  # Coleta as notas dos atletas dadas por cada juiz
            break  # Saida do loop
        elif Opcao == "0":  # Se o usu√°rio quiser voltar
            print("Voltando a sess√£o anterior...")  # Mostra ao usuario que a mensagem do print
            break  # Saida do loop
        else:  # Se a op√ß√£o for inv√°lida ser√° mostrado a mensagem abaixo
            print("\nOp√ß√£o inv√°lida! Tente novamente.\n")  # Mostra ao usuario que a op√ß√£o escolhida fui inv√°lida, e solicita nova entrada de dados

def ColetarNomesJuizes(NumJuizes):  # Fun√ß√£o para coletar os nomes dos ju√≠zes
    print("\n\033[1m===== <<< Coleta de Nomes dos Ju√≠zes >>> =====\033[m")  # Exibe o cabe√ßalho na cor vermelha
    Juizes = []  # Cria uma lista vazia para armazenar os nomes dos ju√≠zes
    for j in range(1, NumJuizes + 1):  # Fun√ß√£o para gerar a sequencia ju√≠zes
        NomeJuiz = input(f"Digite o nome do juiz {j}: ").title()  # Coleta o nome do juiz
        Juizes.append(NomeJuiz)  # Adiciona o nome √† lista de ju√≠zes
    return Juizes  # Retorna a lista de ju√≠zes

CriteriosAvaliacao = {  # Fun√ß√£o de dicion√°rio que armazena os crit√©rios de avalia√ß√£o e suas respectivas pontua√ß√µes
    'Execu√ß√£o': 20,  # Pontua√ß√£o para a execu√ß√£o da manobra com m√°ximo de 20 pontos
    'Dificuldade da Manobra': 20,  # Pontua√ß√£o para a dificuldade da manobra com m√°ximo de 20 pontos
    'Originalidade': 20,  # Pontua√ß√£o para a originalidade da manobra com m√°ximo de 20 pontos
    'Estilo': 20,  # Pontua√ß√£o para estilo da manobra com m√°ximo de 20 pontos
    'Uso da Rampa': 20  # Pontua√ß√£o para o uso da rampa na manobra com m√°ximo de 20 pontos
}

def ColetarNotaCriterio(NomeJuiz, NomeAtleta, Criterio):  # Fun√ß√£o para coletar as notas de cada crit√©rio
    Nota = float(input(f"\nNota do juiz \033[1;33m{NomeJuiz}\033[m para {Criterio} do atleta \033[1;32m{NomeAtleta}\033[m (0 a 20): "))  # Solicita ao juiz que insira a nota para o crit√©rio espec√≠fico dos atletas da competi√ß√£o
    while Nota < 0 or Nota > 20:  # Verifica se a nota est√° fora do intervalo permitido que √© de 0 a 20
        Nota = float(input(f"\033[31mNota inv√°lida. Insira uma nota entre 0 e 20: \033[m")) # Solicita novamente uma nota v√°lida caso a entrada anterior seja inv√°lida
    return Nota  # Retorna a nota v√°lida coletada

def CalcularNotaJuiz(NotasCriterios):  # Fun√ß√£o para calcular a nota final do juiz com base nos crit√©rios
    return sum(NotasCriterios.values())  # Calcula a soma das notas de todos os crit√©rios fornecidos no dicion√°rio NotasCriterios

def CalcularNotaVolta(NotasJuizes):  # Fun√ß√£o para calcular a nota final removendo a maior e a menor nota
    NotasJuizes.sort()  # Ordenar as notas
    NotasFiltradas = NotasJuizes[1:-1]  # Eliminar a menor e a maior nota
    return sum(NotasFiltradas)  # Somar as tr√™s notas restantes

def ColetarNotas(NumAtletas, Juizes):  # Fun√ß√£o para coletar as notas dos atletas de acordo com o ju√≠zes
    Resultados = []  # Lista para armazenar os resultados dos atletas

    for i in range(1, NumAtletas + 1):  # Loop para gerar a sequencia de dados dos atletas
        print(f"\n\033[1m===== <<< Informa√ß√µes do Atleta {i} >>> =====\033[m")  # Cabe√ßalho para as informa√ß√µes do atleta
        NomeAtleta = input("Digite o nome do atleta: ").title()  # Solicita o nome do atleta em formato de t√≠tulo
        PaisAtleta = input(f"Digite o pa√≠s do atleta {NomeAtleta}: ").title()  # Solicita o pa√≠s do atleta em formato de t√≠tulo

        AtletaInfo = {  # Dicion√°rio para armazenar as notas de cada volta do atleta
            "nome": NomeAtleta,  # Armazena o nome do atleta
            "pais": PaisAtleta,  # Armazena o pa√≠s do atleta
            "volta1": 0,  # Inicia a nota da primeira volta em 0
            "volta2": 0,  # Inicia a nota da segunda volta em 0
            "volta3": 0   # Inicia a nota da terceira volta em 0
        }

        for j in range(1, 3 + 1):  # Loop para informar sobre as voltas 1, 2 e 3
            print(f"\n\033[1m===== <<< Notas para a Volta {j} do Atleta \033[1;32m{NomeAtleta}\033[m >>> =====\033[m")  # Cabe√ßalho para as notas da volta atual
            NotasVolta = []  # Func√ß√£o de lista para armazenar as notas dos ju√≠zes para essa volta

            for NomeJuiz in Juizes:  # Loop para cada juiz usando a lista de ju√≠zes coletada
                print(f"\n      \033[1m<<< Avalia√ß√£o do Juiz: \033[1;33m{NomeJuiz}\033[m >>>      \033[m")  # Exibir cabe√ßalho das notas antes do primeiro crit√©rio
                NotasCriterios = {}  # Lista para armazenar as notas da volta atual

                for Criterio in CriteriosAvaliacao.keys(): # Coletar as notas para cada crit√©rio
                    Nota = ColetarNotaCriterio(NomeJuiz, NomeAtleta, Criterio)  # Coleta a nota do juiz para o crit√©rio
                    NotasCriterios[Criterio] = Nota  # Armazenamento da nota no dicion√°rio de crit√©rios

                NotaFinalJuiz = CalcularNotaJuiz(NotasCriterios)  # Calcula a nota final do juiz somando as notas dos crit√©rios
                NotasVolta.append(NotaFinalJuiz)  # Adiciona a nota final do juiz √† lista de notas da volta atual

                # Exibir a nota final com a nova formata√ß√£o
                print(f"\n\033[1m------------->>> Nota final do atleta \033[1;32m{NomeAtleta}\033[m ({PaisAtleta}) dada pelo juiz \033[1;33m{NomeJuiz}\033[m: {NotaFinalJuiz:.2f} <<<-------------\033[m")

            NotaFinalVolta = CalcularNotaVolta(NotasVolta)  # Calcular a pontua√ß√£o da volta eliminando a maior e a menor nota
            AtletaInfo[f"volta{j}"] = NotaFinalVolta  # Armazenar a nota final da volta no dicion√°rio
            print(f"\n\033[1m------------->>> Pontua√ß√£o Total da Volta {j} do Atleta \033[1;32m{NomeAtleta}\033[m: {NotaFinalVolta:.2f} pontos <<<-------------\033[m")  # Exibe a pontua√ß√£o total da volta

        AtletaInfo["nota_final"] = max(AtletaInfo["volta1"], AtletaInfo["volta2"], AtletaInfo["volta3"])  # Determinar a maior pontua√ß√£o entre as voltas como a nota final do atleta

        Resultados.append(AtletaInfo)  # Armazenar os resultados do atleta

    ExibirPodio(Resultados)  # Exibir o p√≥dio

def ExibirPodio(Resultados):  # Fun√ß√£o para exibir o p√≥dio dos melhores atletas

    ResultadosOrdenados = sorted(Resultados, key=lambda x: x['nota_final'], reverse=True)  # Ordenar os resultados pela pontua√ß√£o em ordem decrescente

    print("\n    \033[1mP√≥dio dos 3 Melhores Atletas:\033[m")  # Exibe o p√≥dio
    if len(ResultadosOrdenados) > 0:  # Verifica se h√° resultados ordenados para exibir
        print("\n                      \033[1mü•á 1¬∫ Lugar\033[m")  # Exibe o t√≠tulo do 1¬∫ lugar
        print(f"                  {ResultadosOrdenados[0]['nome']} - {ResultadosOrdenados[0]['nota_final']:.2f} ({ResultadosOrdenados[0]['pais']})")  # Exibe o nome, nota e pa√≠s do 1¬∫ atleta
    if len(ResultadosOrdenados) > 1:  # Verifica se h√° pelo menos 2 atletas para exibir o 2¬∫ e 3¬∫ lugares
        print("\n      \033[1mü•à 2¬∫ Lugar                       ü•â 3¬∫ Lugar\033[m")  # Exibe os t√≠tulos para o 2¬∫ e 3¬∫ lugares
        print(f"    {ResultadosOrdenados[1]['nome']} - {ResultadosOrdenados[1]['nota_final']:.2f} ({ResultadosOrdenados[1]['pais']})              {ResultadosOrdenados[2]['nome']} - {ResultadosOrdenados[2]['nota_final']:.2f} ({ResultadosOrdenados[2]['pais']})")  # Exibe nome, nota e pa√≠s dos 2¬∫ e 3¬∫ atletas

    # Exibir a tabela com todos os resultados
    print("\nResultados Gerais:")  # T√≠tulo para a se√ß√£o de resultados gerais
    print(f"\033[1m \n{'Rank':<10} {'Nome':<20} {'Pa√≠s':<20} {'Volta 1':<10} {'Volta 2':<10} {'Volta 3':<10} {'Pontua√ß√£o final':<10}\033[m")  # Cabe√ßalho da tabela com formata√ß√£o
    print("-" * 100)  # Linha separadora
    for rank, atleta in enumerate(ResultadosOrdenados, start=1):  # Loop para exibir cada atleta com seu rank e resultados
        print(f"{rank:<10} {atleta['nome']:<20} {atleta['pais']:<20} {atleta['volta1']:<10.2f} {atleta['volta2']:<10.2f} {atleta['volta3']:<10.2f} {atleta['nota_final']:<10.2f}")  # Exibe os detalhes dos atletas formatado em rank

# Iniciar o programa
NotaSkate()
```
