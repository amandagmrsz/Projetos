```python
def NotaSkate():  # Função que inicia o programa de notas da competição
    while True:  # Tem a função de loop de escolha de opção, até o usuario optar por sair
        print("\n          ===== <<< MODALIDADE >>>  =====")
        print("         |         \033[31m[1]\033[m Street            |")  # Opção de modalidade exibidas
        print("         |         \033[31m[2]\033[m Park              |")
        print("         |         \033[31m[0]\033[m Sair              |")  # Opção do usuario sair do programa
        print("         --------------------------------")
        Opcao = input("Insira a opção: ")  # Solicita ao usuario inserir uma opção

        if Opcao == "1":  # Se a opção for 1, escolheu Street
            print("\nVocê escolheu a opção: Street\n")
            Sexo()  # Chama a função Sexo para coletar o sexo do atleta
        elif Opcao == "2":  # Se a opção for 2, escolheu Park
            print("\nVocê escolheu a opção: Park\n")
            Sexo()  # Chama a função Sexo para coletar o sexo do atleta
        elif Opcao == "0":  # Se a opção for 0, o usuario deseja sair
            print("Saindo do sistema...")
            break  # Sai do loop e finaliza o programa
        else:  # Se a opção for inválida será mostrado a mensagem abaixo
            print("\nOpção inválida! Tente novamente.\n")  # Mostra ao usuario que a opção escolhida fui inválida, e solicita nova entrada de dados

def Sexo():  # Função para escolher o sexo dos atletas
    while True:  # Inicia um loop para coletar o sexo
        print("\n          ===== <<< Sexo >>> =====")
        print("         |    \033[31m[1]\033[m Feminino        |")  # Opção de sexo do atleta exibidas
        print("         |    \033[31m[2]\033[m Masculino       |")
        print("         |    \033[31m[0]\033[m Voltar          |")  # Opção do usuario sair do programa
        print("         --------------------------")
        Opcao = input("Insira a opção: ")  # Solicita ao usuario inserir uma opção

        if Opcao in ["1", "2"]:  # Se a opção for 1 ou 2, o usuario escolheu o sexo do atleta
            if Opcao == "1":  # Se a opção for 1, escolheu o sexo Feminino
                print("\nVocê escolheu a opção: Feminino\n") # Informa a opção escolhida
            else:  # Se a opção for 2, escolheu o sexo Masculino
                print("\nVocê escolheu a opção: Masculino\n")  # Informa a opção escolhida
            print("\n===== <<<.............................. >>> =====")
            NumAtletas = int(input("Quantos atletas participarão? "))  # Solicita ao usuario a quantidade de atletas participantes da competição
            print("\n===== <<<.............................. >>> =====")
            NumJuizes = 5  # Define que sempre serão 5 juízes
            Juizes = ColetarNomesJuizes(NumJuizes)  # Coleta os nomes dos juízes
            ColetarNotas(NumAtletas, Juizes)  # Coleta as notas dos atletas dadas por cada juiz
            break  # Saida do loop
        elif Opcao == "0":  # Se o usuário quiser voltar
            print("Voltando a sessão anterior...")  # Mostra ao usuario que a mensagem do print
            break  # Saida do loop
        else:  # Se a opção for inválida será mostrado a mensagem abaixo
            print("\nOpção inválida! Tente novamente.\n")  # Mostra ao usuario que a opção escolhida fui inválida, e solicita nova entrada de dados

def ColetarNomesJuizes(NumJuizes):  # Função para coletar os nomes dos juízes
    print("\n\033[1m===== <<< Coleta de Nomes dos Juízes >>> =====\033[m")  # Exibe o cabeçalho na cor vermelha
    Juizes = []  # Cria uma lista vazia para armazenar os nomes dos juízes
    for j in range(1, NumJuizes + 1):  # Função para gerar a sequencia juízes
        NomeJuiz = input(f"Digite o nome do juiz {j}: ").title()  # Coleta o nome do juiz
        Juizes.append(NomeJuiz)  # Adiciona o nome à lista de juízes
    return Juizes  # Retorna a lista de juízes

CriteriosAvaliacao = {  # Função de dicionário que armazena os critérios de avaliação e suas respectivas pontuações
    'Execução': 20,  # Pontuação para a execução da manobra com máximo de 20 pontos
    'Dificuldade da Manobra': 20,  # Pontuação para a dificuldade da manobra com máximo de 20 pontos
    'Originalidade': 20,  # Pontuação para a originalidade da manobra com máximo de 20 pontos
    'Estilo': 20,  # Pontuação para estilo da manobra com máximo de 20 pontos
    'Uso da Rampa': 20  # Pontuação para o uso da rampa na manobra com máximo de 20 pontos
}

def ColetarNotaCriterio(NomeJuiz, NomeAtleta, Criterio):  # Função para coletar as notas de cada critério
    Nota = float(input(f"\nNota do juiz \033[1;33m{NomeJuiz}\033[m para {Criterio} do atleta \033[1;32m{NomeAtleta}\033[m (0 a 20): "))  # Solicita ao juiz que insira a nota para o critério específico dos atletas da competição
    while Nota < 0 or Nota > 20:  # Verifica se a nota está fora do intervalo permitido que é de 0 a 20
        Nota = float(input(f"\033[31mNota inválida. Insira uma nota entre 0 e 20: \033[m")) # Solicita novamente uma nota válida caso a entrada anterior seja inválida
    return Nota  # Retorna a nota válida coletada

def CalcularNotaJuiz(NotasCriterios):  # Função para calcular a nota final do juiz com base nos critérios
    return sum(NotasCriterios.values())  # Calcula a soma das notas de todos os critérios fornecidos no dicionário NotasCriterios

def CalcularNotaVolta(NotasJuizes):  # Função para calcular a nota final removendo a maior e a menor nota
    NotasJuizes.sort()  # Ordenar as notas
    NotasFiltradas = NotasJuizes[1:-1]  # Eliminar a menor e a maior nota
    return sum(NotasFiltradas)  # Somar as três notas restantes

def ColetarNotas(NumAtletas, Juizes):  # Função para coletar as notas dos atletas de acordo com o juízes
    Resultados = []  # Lista para armazenar os resultados dos atletas

    for i in range(1, NumAtletas + 1):  # Loop para gerar a sequencia de dados dos atletas
        print(f"\n\033[1m===== <<< Informações do Atleta {i} >>> =====\033[m")  # Cabeçalho para as informações do atleta
        NomeAtleta = input("Digite o nome do atleta: ").title()  # Solicita o nome do atleta em formato de título
        PaisAtleta = input(f"Digite o país do atleta {NomeAtleta}: ").title()  # Solicita o país do atleta em formato de título

        AtletaInfo = {  # Dicionário para armazenar as notas de cada volta do atleta
            "nome": NomeAtleta,  # Armazena o nome do atleta
            "pais": PaisAtleta,  # Armazena o país do atleta
            "volta1": 0,  # Inicia a nota da primeira volta em 0
            "volta2": 0,  # Inicia a nota da segunda volta em 0
            "volta3": 0   # Inicia a nota da terceira volta em 0
        }

        for j in range(1, 3 + 1):  # Loop para informar sobre as voltas 1, 2 e 3
            print(f"\n\033[1m===== <<< Notas para a Volta {j} do Atleta \033[1;32m{NomeAtleta}\033[m >>> =====\033[m")  # Cabeçalho para as notas da volta atual
            NotasVolta = []  # Funcção de lista para armazenar as notas dos juízes para essa volta

            for NomeJuiz in Juizes:  # Loop para cada juiz usando a lista de juízes coletada
                print(f"\n      \033[1m<<< Avaliação do Juiz: \033[1;33m{NomeJuiz}\033[m >>>      \033[m")  # Exibir cabeçalho das notas antes do primeiro critério
                NotasCriterios = {}  # Lista para armazenar as notas da volta atual

                for Criterio in CriteriosAvaliacao.keys(): # Coletar as notas para cada critério
                    Nota = ColetarNotaCriterio(NomeJuiz, NomeAtleta, Criterio)  # Coleta a nota do juiz para o critério
                    NotasCriterios[Criterio] = Nota  # Armazenamento da nota no dicionário de critérios

                NotaFinalJuiz = CalcularNotaJuiz(NotasCriterios)  # Calcula a nota final do juiz somando as notas dos critérios
                NotasVolta.append(NotaFinalJuiz)  # Adiciona a nota final do juiz à lista de notas da volta atual

                # Exibir a nota final com a nova formatação
                print(f"\n\033[1m------------->>> Nota final do atleta \033[1;32m{NomeAtleta}\033[m ({PaisAtleta}) dada pelo juiz \033[1;33m{NomeJuiz}\033[m: {NotaFinalJuiz:.2f} <<<-------------\033[m")

            NotaFinalVolta = CalcularNotaVolta(NotasVolta)  # Calcular a pontuação da volta eliminando a maior e a menor nota
            AtletaInfo[f"volta{j}"] = NotaFinalVolta  # Armazenar a nota final da volta no dicionário
            print(f"\n\033[1m------------->>> Pontuação Total da Volta {j} do Atleta \033[1;32m{NomeAtleta}\033[m: {NotaFinalVolta:.2f} pontos <<<-------------\033[m")  # Exibe a pontuação total da volta

        AtletaInfo["nota_final"] = max(AtletaInfo["volta1"], AtletaInfo["volta2"], AtletaInfo["volta3"])  # Determinar a maior pontuação entre as voltas como a nota final do atleta

        Resultados.append(AtletaInfo)  # Armazenar os resultados do atleta

    ExibirPodio(Resultados)  # Exibir o pódio

def ExibirPodio(Resultados):  # Função para exibir o pódio dos melhores atletas

    ResultadosOrdenados = sorted(Resultados, key=lambda x: x['nota_final'], reverse=True)  # Ordenar os resultados pela pontuação em ordem decrescente

    print("\n    \033[1mPódio dos 3 Melhores Atletas:\033[m")  # Exibe o pódio
    if len(ResultadosOrdenados) > 0:  # Verifica se há resultados ordenados para exibir
        print("\n                      \033[1m🥇 1º Lugar\033[m")  # Exibe o título do 1º lugar
        print(f"                  {ResultadosOrdenados[0]['nome']} - {ResultadosOrdenados[0]['nota_final']:.2f} ({ResultadosOrdenados[0]['pais']})")  # Exibe o nome, nota e país do 1º atleta
    if len(ResultadosOrdenados) > 1:  # Verifica se há pelo menos 2 atletas para exibir o 2º e 3º lugares
        print("\n      \033[1m🥈 2º Lugar                       🥉 3º Lugar\033[m")  # Exibe os títulos para o 2º e 3º lugares
        print(f"    {ResultadosOrdenados[1]['nome']} - {ResultadosOrdenados[1]['nota_final']:.2f} ({ResultadosOrdenados[1]['pais']})              {ResultadosOrdenados[2]['nome']} - {ResultadosOrdenados[2]['nota_final']:.2f} ({ResultadosOrdenados[2]['pais']})")  # Exibe nome, nota e país dos 2º e 3º atletas

    # Exibir a tabela com todos os resultados
    print("\nResultados Gerais:")  # Título para a seção de resultados gerais
    print(f"\033[1m \n{'Rank':<10} {'Nome':<20} {'País':<20} {'Volta 1':<10} {'Volta 2':<10} {'Volta 3':<10} {'Pontuação final':<10}\033[m")  # Cabeçalho da tabela com formatação
    print("-" * 100)  # Linha separadora
    for rank, atleta in enumerate(ResultadosOrdenados, start=1):  # Loop para exibir cada atleta com seu rank e resultados
        print(f"{rank:<10} {atleta['nome']:<20} {atleta['pais']:<20} {atleta['volta1']:<10.2f} {atleta['volta2']:<10.2f} {atleta['volta3']:<10.2f} {atleta['nota_final']:<10.2f}")  # Exibe os detalhes dos atletas formatado em rank

# Iniciar o programa
NotaSkate()
```
