**Desafio 2.3 – INDIVIDUAL **

Foi feita uma pesquisa em uma universidade entre os acadêmicos matriculados a instituição. Todos os acadêmicos que estavam presentes no dia da pesquisa, preencheram o seguinte questionário:


Curso: (C - Ciências da computação (CCP); A - Analise e desenvolvimento de sistemas (ADS))

Idade: (em anos)

Gosta do curso que está fazendo: (S - sim; N - não)



O número de pessoas que responderam ao questionário é indeterminado.

Faça um CÓDIGO EM PY que seja capaz de calcular e informar:

●O número de alunos de CCP com menos de 23 anos;

●O número de alunos de ADS que estão gostando do curso que estão fazendo;

●O total de alunos entrevistados.

```python
#Inicialização das variáveis
TotalAlunos = 0
CcpMenos23 = 0
AdsGostandoCurso = 0

#Loop para entrada de dados
while True:
    NomeAluno = input("Digite seu NOME ou digite Q para sair: ").upper()

    if NomeAluno == "Q":
        break


    Curso = input("Curso (C para Ciências da Computação, A para ADS): ").upper()

    Idade = int(input("Idade do aluno: "))
    GostandoCurso = input("Gosta do curso (S para Sim, N para Não) ").upper()

    print()  # Adiciona uma linha em branco

    #Contagem total de alunos
    TotalAlunos += 1

    #Contagem de alunos de CCP com menos de 23 anos
    if Curso == "C" and Idade < 23:
        CcpMenos23 += 1
    #Contagem de alunos de ADS que gostam do curso
    if Curso == "A" and GostandoCurso == "S":
        AdsGostandoCurso += 1

#Exibindo os resultados
print("\n\033[1mResultados da Entrevista:")
print(f"Total de alunos de CCP com menos de 23 anos: {CcpMenos23}")
print(f"Total de alunos de ADS que gostam do curso: {AdsGostandoCurso}")
print(f"Total de alunos entrevistados: {TotalAlunos}")
```

 *Desafio 2.4 – INDIVIDUAL*

Elabore um CÓDIGO EM PY que simule a gestão de finanças pessoais. O algoritmo deverá:

1.Solicitar ao usuário o valor inicial da receita.

2.Permitir que o usuário insira operações financeiras (débito ou crédito) repetidamente, onde:

○Se a operação for "crédito", o valor deve ser somado à receita.

○Se a operação for "débito", o valor deve ser subtraído da receita.

○Se a operação for “concluir”, o algoritmo deve exibir o extrato.

3.Ao final, o CODIGO deverá exibir um extrato contendo:

○O valor inicial da receita.

○O total de débitos.

○O total de créditos.

○O valor final da receita após todas as operações.



Lembre-se de usar estruturas de repetição para permitir múltiplas operações.

```python
#Solicita o nome do usuário e o valor inicial da receita
Nome = input("Digite seu nome: ")
ReceitaInicial = float(input("Digite o valor inicial da receita: "))

#Solicita o mês para o extrato financeiro
Mes = input("Digite o mês referente ao extrato: ")

#Variáveis para armazenar o total de créditos, débitos e a receita final
TotalCreditos = 0
TotalDebitos = 0
ReceitaFinal = ReceitaInicial

#Loop para permitir a inserção de múltiplas operações financeiras
while True:
    operacao = input("Digite a operação (crédito/débito/concluir): ").lower()

    if operacao == "crédito":
        valor = float(input("Digite o valor do crédito: "))
        ReceitaFinal += valor
        TotalCreditos += valor
    elif operacao == "débito":
        valor = float(input("Digite o valor do débito: "))
        ReceitaFinal -= valor
        TotalDebitos += valor
    elif operacao == "concluir":
        break
    else:
        print("Operação inválida. Tente novamente.")

#Exibição do extrato financeiro
print(f"\n\033[1m{Nome}, seu extrato financeiro do mês de {Mes} é:\033[m")
print(f"Valor inicial da receita: R$ {ReceitaInicial:.2f}")
print(f"Total de créditos: R$ {TotalCreditos:.2f}")
print(f"Total de débitos: R$ {TotalDebitos:.2f}")
print(f"Valor final da receita: R$ {ReceitaFinal:.2f}")
```

*Desafio 2.5 – INDIVIDUAL*

Desenvolva um código em python para um aplicativo de pizzaria, que permite que o usuário faça um pedido de pizza seguindo os passos abaixo:

●Escolha do tamanho da pizza: O usuário deve escolher entre os tamanhos disponíveis (por exemplo, pequena, média, grande). Cada tamanho deverá conter um fator de multiplicação (Ex.: Pequena = 1; Média = 1,2; Grande = 1,5)

●Escolha dos sabores: O usuário deve escolher entre um ou dois sabores de pizza.

●Seleção dos sabores: O usuário deve selecionar os sabores desejados (Ex.: Calabresa, Marguerita, Quatro queijos e Frango).

●Permitir que o usuário peça mais de uma pizza.

Na saída, o código deve:

●Mostrar o valor total do pedido: Calcule o valor da pizza com base no tamanho, sabores escolhidos e quantidade solicitada.

●Mostrar opções de finalização:

○Gostaria de fazer um novo pedido (S - Sim ou N - Não)?

```python
# Definição de preços básicos para pizza pequena por sabor
SaboresPrecos = {
    "Calabresa": 25.0,
    "Marguerita": 24.5,
    "Quatro Queijos": 28.0,
    "Frango": 20.0
}

# Fatores de multiplicação para tamanhos
Tamanhos = {
    "Pequena": 1,
    "Média": 1.2,
    "Grande": 1.5
}

def CalcularValorPizza(Tamanho, Sabores):
    ValorBase = 0
    for Sabor in Sabores:
        ValorBase += SaboresPrecos[Sabor]
    return ValorBase * Tamanhos[Tamanho]


def PedidoPizzaria():
    TotalPedido = 0
    Pedidos = []


    print("\033[1mBem-vindo à Pizzaria!\033[m")
    # Solicitar o nome do cliente
    NomeCliente = input("Digite o nome do cliente: ")

    while True:
        # Escolha do tamanho da pizza
        print("\nEscolha o tamanho da pizza:")
        print("1. Pequena")
        print("2. Média")
        print("3. Grande")
        TamanhoOpcao = input("Digite o número do tamanho desejado: ")

        if TamanhoOpcao == "1":
            Tamanho = "Pequena"
        elif TamanhoOpcao == "2":
            Tamanho = "Média"
        elif TamanhoOpcao == "3":
            Tamanho = "Grande"
        else:
            print("Tamanho inválido! Escolha novamente.")
            continue

        # Escolha dos sabores
        Sabores = []
        print("\nEscolha o(s) sabor(es) da pizza (até 2 sabores):")
        for i, Sabor in enumerate(SaboresPrecos.keys(), 1):
            print(f"{i}. {Sabor}")
        Sabor1 = input("Digite o número do primeiro sabor: ")
        Sabores.append(list(SaboresPrecos.keys())[int(Sabor1) - 1])

        Sabor2 = input("Gostaria de adicionar um segundo sabor? (Digite o número ou pressione Enter para pular): ")
        if Sabor2:
            Sabores.append(list(SaboresPrecos.keys())[int(Sabor2) - 1])

        # Quantidade de pizzas
        Quantidade = int(input("Digite a quantidade de pizzas desejada: "))

        # Calcular o valor da pizza
        ValorPizza = CalcularValorPizza(Tamanho, Sabores)
        ValorTotal = ValorPizza * Quantidade
        TotalPedido += ValorTotal

        # Salvar detalhes do pedido
        Pedidos.append({
            "Tamanho": Tamanho,
            "Sabores": Sabores,
            "Quantidade": Quantidade,
            "ValorPizza": ValorPizza,
            "ValorTotal": ValorTotal
        })

        print(f"\nValor total do pedido atual: R$ {TotalPedido:.2f}")

        # Verificar se o cliente quer fazer um novo pedido
        NovoPedido = input("Gostaria de fazer um novo pedido? (S - Sim ou N - Não): ").lower()
        if NovoPedido != 's':
            break

    # Exibir nota fiscal no final
    print()
    print(f"\n{'='*30} NOTA FISCAL {'='*30}")
    print(f"Cliente: {NomeCliente}")
    print(f"{'-'*70}")
    for NumeroPedido, Pedido in enumerate(Pedidos, 1):
        print(f"Pedido {NumeroPedido}:")
        print(f"  Tamanho: {Pedido['Tamanho']}")
        print(f"  Sabores: {', '.join(Pedido['Sabores'])}")
        print(f"  Quantidade: {Pedido['Quantidade']}")
        print(f"  Valor unitário: R$ {Pedido['ValorPizza']:.2f}")
        print(f"  Subtotal: R$ {Pedido['ValorTotal']:.2f}")
        print(f"{'-'*70}")

    print(f"\nValor total do pedido final: R$ {TotalPedido:.2f}")
    print(f"{'='*70}")

# Executar o programa
PedidoPizzaria()
```
