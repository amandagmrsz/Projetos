```python
def MenuFinanceiro():
    while True:
        print("\n         ===== <<< FINANCEIRO >>> =====")
        print("         | [1] Monitoramento de Gastos |")
        print("         | [2] Holerite                |")
        print("         | [3] Investimentos           |")
        print("         | [0] Sair                    |")
        print("         ------------------------------")
        opcao = input("Insira a opção: ")

        if opcao == "1":
            print("\nVocê escolheu a opção: Monitoramento de Gastos\n")
            MonitoramentoGastos()
        elif opcao == "2":
            print("\nVocê escolheu a opção: Holerite\n")
            Holerite()
        elif opcao == "3":
            print("\nVocê escolheu a opção: Investimentos\n")
            SimuladorInvestimentos()
        elif opcao == "0":
            print("Saindo do sistema...")
            break
        else:
            print("\nOpção inválida! Tente novamente.\n")

def SimuladorInvestimentos():
    Nome = str(input("Qual seu nome: ")).title()
    IdadeAtual = int(input("Qual sua idade atual: "))

    print("\nEscolha o nível de risco para seus investimentos:")
    print("1. Baixo Risco (SELIC)")
    print("2. Médio Risco (CDB)")
    print("3. Alto Risco (BITCOIN)")
    Risco = input("Digite o número da opção escolhida: ")

    if Risco == "1":
        TaxaAnuais = 0.13  # Aproximadamente 13% ao ano para a SELIC
        Investimentos = [
            {"nome": "SELIC", "tipo": "Taxa Referencial", "rentabilidade": 13.0, "tributacao": "Isento"}
        ]
    elif Risco == "2":
        TaxaAnuais = 0.12  # Aproximadamente 12% ao ano para o CDB
        Investimentos = [
            {"nome": "CDB", "tipo": "Certificado de Depósito Bancário", "rentabilidade": 12.0, "tributacao": "Imposto de Renda Progressivo (até 22.5%)"}
        ]
    elif Risco == "3":
        TaxaAnuais = 0.150  # Aproximadamente 150% ao ano para o Bitcoin (muito volátil)
        Investimentos = [
            {"nome": "BITCOIN", "tipo": "Criptomoeda", "rentabilidade": 150.0, "tributacao": "Imposto de Renda sobre ganhos acima de R$ 35.000"}
        ]
    else:
        print("Opção inválida!")
        return  # Termina a função sem continuar

    # Imprime o cabeçalho da tabela
    print(f"\n{'Nome do Investimento':<25}{'Tipo de Investimento':<35}{'Rentabilidade (%)':<20}{'Imposto'}")
    print("-" * 90)

    # Imprime os dados de cada investimento a partir da lista armazenada
    for investimento in Investimentos:
        print(f"{investimento['nome']:<25}{investimento['tipo']:<35}{investimento['rentabilidade']:<20.2f}{investimento['tributacao']}")

    ValorMensalInvestido = float(input("\nQuanto você pode investir por mês? R$: "))
    Meses = int(input("Por quantos meses você pretende investir? "))

    TaxaMensal = TaxaAnuais / 12
    MontanteFinal = 0
    RendimentoMensal = []

    for x in range(Meses):
        MontanteFinal = (MontanteFinal + ValorMensalInvestido) * (1 + TaxaMensal)
        RendimentoMensal.append(MontanteFinal - (ValorMensalInvestido * (x + 1)))  # Rendimento de cada mês

    TotalInvestido = ValorMensalInvestido * Meses
    RendimentoTotal = MontanteFinal - TotalInvestido

    # Exibe os resultados mensais
    print(f"\nOlá, {Nome}! Aqui estão os resultados da sua simulação:")

    # Exibe o rendimento mensal
    print(f"\nRendimento Mensal de Cada Mês:")
    print(f"{'Mês':<10}{'Rendimento':<15}")
    for i, rendimento in enumerate(RendimentoMensal, 1):
        print(f"{i:<10}{rendimento:<15.2f}")

    print(f"\nValor total investido: R$ {TotalInvestido:.2f}")
    print(f"Valor final após {Meses} meses: R$ {MontanteFinal:.2f}")
    print(f"Rendimento total: R$ {RendimentoTotal:.2f}")


def MonitoramentoGastos():
    import pandas as pd
    # Lista de meses
    meses = ["Janeiro", "Fevereiro", "Março", "Abril", "Maio", "Junho",
            "Julho", "Agosto", "Setembro", "Outubro", "Novembro", "Dezembro"]

    # Estrutura de categorias e subcategorias financeiras
    dados = {
        "RECEITAS": ["Salário", "Aluguel", "Pensão", "Horas extras", "13º salário", "Férias", "Rendimentos", "Outros"],
        "INVESTIMENTOS": ["Ações", "Tesouro Direto", "Renda fixa", "Previdência privada", "Outros"],

        "DESPESAS FIXAS": ["Habitação", "Transporte", "Saúde", "Educação", "Imposto", "Outros"],
        "Habitação": ["Aluguel", "Condomínio", "Prestação da casa", "Seguro da casa", "Diarista"],
        "Transporte": ["Pestação do carro", "Seguro do Carro"],
        "Saúde": ["Plano Saúde", "Seguro Saúde"],
        "Educação": ["Colégio", "Faculdade", "Cursos"],
        "Imposto": ["IPTU", "IPVA"],
        "Outros": ["Empréstimos", "Seguro de Vida"],

        "DESPESAS VARIÁVEIS": ["Habitação", "Transporte", "Alimentação", "Saúde", "Cuidados Pessoais"],
        "Habitação": ["Luz", "Água", "Telefone", "Gás", "Tv e Streaming", "Internet"],
        "Transporte": ["Metrô", "Ônibus", "Uber e Táxi", "Combustível", "Estacionamento"],
        "Alimentação": ["Super Mercado", "Feira", "Padaria", "Delivery"],
        "Saúde": ["Medicamentos"],
        "Cuidados Pessoais": ["Cabelereiro", "Manicure", "Esteticista", "Academia", "Clube"],

        "DESPESAS EXTRAS": ["Médico", "Dentista", "Hospital","Carro (manutenção)", "Casa (manutenção)", "Material escolar", "Uniforme"],
        "DESPESAS ADICIONAIS": ["Viagens", "Cinema/Tratro", "Roupas", "Calçados", "Restaurantes/Bares", "Presentes", "Acessórios", "Outros"]
    }

    # Criar DataFrames para cada categoria com subcategorias e meses
    tabelas = {}
    for categoria, subcategorias in dados.items():
        tabelas[categoria] = pd.DataFrame({
            "Subcategoria": subcategorias,
            **{mes: [0] * len(subcategorias) for mes in meses},  # Inicializa com zeros
            "Total Anual": [0] * len(subcategorias),
            "Média Mensal": [0] * len(subcategorias),
        })

    # Função para calcular totais e médias
    def calcular_totais(df):
        df["Total Anual"] = df[meses].sum(axis=1)
        df["Média Mensal"] = df[meses].mean(axis=1)
        return df

    # Preenchimento de valores por mês
    def preencher_mes(tabelas, mes_idx):
        mes_atual = meses[mes_idx]
        print(f"\nPreenchendo dados para o mês de {mes_atual}:")
        for categoria, tabela in tabelas.items():
            print(f"\nCategoria: {categoria}")
            for i, row in tabela.iterrows():
                subcategoria = row["Subcategoria"]
                while True:
                    try:
                        valor = float(input(f"Digite o valor para {subcategoria} ({categoria}): "))
                        tabelas[categoria].at[i, mes_atual] = valor
                        break
                    except ValueError:
                        print("Por favor, insira um valor numérico válido.")
        return tabelas

    # Cálculo do saldo total (mensal ou anual)
    def calcular_saldos(tabelas):
        total_receitas = tabelas["RECEITAS"]["Total Anual"].sum()
        total_investimentos = tabelas["INVESTIMENTOS"]["Total Anual"].sum()
        total_despesas = sum(
            tabelas[cat]["Total Anual"].sum()
            for cat in ["DESPESAS FIXAS", "DESPESAS VARIÁVEIS", "DESPESAS EXTRAS", "DESPESAS ADICIONAIS"]
        )
        saldo = total_receitas - (total_investimentos + total_despesas)
        return total_receitas, total_investimentos, total_despesas, saldo

    # Preencher os dados mês a mês
    for mes_idx in range(len(meses)):
        tabelas = preencher_mes(tabelas, mes_idx)
        for categoria in tabelas:
            tabelas[categoria] = calcular_totais(tabelas[categoria])

        # Resumo mensal
        print("\nResumo do mês:")
        total_receitas, total_investimentos, total_despesas, saldo = calcular_saldos(tabelas)
        print(f"Receitas: R$ {total_receitas:.2f}")
        print(f"Investimentos: R$ {total_investimentos:.2f}")
        print(f"Despesas: R$ {total_despesas:.2f}")
        print(f"Saldo: R$ {saldo:.2f}")

    # Cálculo de totais anuais
    print("\nResumo Anual:")
    total_receitas, total_investimentos, total_despesas, saldo = calcular_saldos(tabelas)
    print(f"Total Receitas: R$ {total_receitas:.2f}")
    print(f"Total Investimentos: R$ {total_investimentos:.2f}")
    print(f"Total Despesas: R$ {total_despesas:.2f}")
    print(f"Saldo Final: R$ {saldo:.2f}")

    # Exibir as tabelas completas
    for categoria, tabela in tabelas.items():
        print(f"\nCategoria: {categoria}")
        print(tabela)

def Holerite():
        # Desenvolvimento de projeto para calculo trabalhista
    # Iniciando com o nome do funcionário
    nome = input("Informe o nome do funcionário(a): ")
    salario_base = float(input("Valor do salário base: "))
    horas_trabalhadas = float(input("Informe a quantidade de horas trabalhadas (mensal): "))
    horas_extras = float(input("Informe a quantidade de horas extras se houver: "))
    vale_transporte = float(input("Valor do vale transporte:R$ "))
    vale_alimentacao = float(input("Valor do vale alimentação:R$ "))
    vale_refeicao = float(input("Valor do vale refeição:R$ "))
    adicional_noturno = float(input("Valor do adicional noturno:R$ "))
    outros_descontos = float(input("Valor de outros descontos:R$ "))

    # Informações do cálculo de acordo com a tabela do INSS
    if salario_base <= 1412:
        inss = salario_base * 0.07
    elif salario_base <= 2666.68:
        inss = salario_base * 0.09
    elif salario_base <= 4000.03:
        inss = salario_base * 0.12
    elif salario_base <= 4000.04:
        inss = salario_base * 0.14
    elif salario_base < 7786.02:
        inss = salario_base * 0.14
    else:
        inss = 7786.02 * 0.14

    # Informações do cálculo de acordo com a tabela do IRRF
    if salario_base <= 2112.01:
        irrf = 0
    elif salario_base <= 2112.01:
        irrf = (salario_base * 0.075) - 158.40
    elif salario_base <= 2826.66:
        irrf = (salario_base * 0.15) - 370.40
    elif salario_base <= 3751.06:
        irrf = (salario_base * 0.225) - 651.73
    elif salario_base < 4664.69:
        irrf = (salario_base * 0.275) - 884.96
    else:
        irrf = 4664.69 * 0.275

    # Informações do cálculo do FGTS
    fgts = salario_base * 0.08

    # Holerite do funcionário
    print(f"\n{'='*32} Holerite {'='*32}")
    print(f"Funcionário: {nome}")
    print(f"{'-'*74}")
    print(f"{'Recebimentos':<35}{'Descontos':>35}")
    print(f"{'-'*74}")
    print(f"{'Salário Base:':<25} R$ {salario_base:<10.2f}{'Desconto INSS:':>20} R$ {inss:>10.2f}")
    print(f"{'Vale Transporte:':<25} R$ {vale_transporte:<10.2f}{'Desconto IRRF:':>20} R$ {irrf:>10.2f}")
    print(f"{'Vale Alimentação:':<25} R$ {vale_alimentacao:<10.2f}{'Outros Descontos:':>20} R$ {outros_descontos:>10.2f}")
    print(f"{'Vale Refeição:':<25} R$ {vale_refeicao:<10.2f}{'':>20} {'':>10}")
    print(f"{'Adicional Noturno:':<25} R$ {adicional_noturno:<10.2f}{'':>20} {'':>10}")
    print(f"{'FGTS (8%):':<25} R$ {fgts:<10.2f}{'':>20} {'':>10}")
    print(f"{'='*74}")
    print(f"{'Salário Líquido:':<25} R$ {(salario_base - inss - irrf + vale_transporte + vale_alimentacao + vale_refeicao + adicional_noturno - outros_descontos + fgts):<10.2f}")
    print(f"{'-'*74}")
    print(f"{'='*742}")

# Chama o menu financeiro
MenuFinanceiro()
```
