```python
def NotaSkate():
    while True:
        print("\n         ===== <<< MODALIDADE >>> =====")
        print("         |         [1] Street            |")
        print("         |         [2] Park              |")
        print("         |         [0] Sair              |")
        print("           ------------------------------")
        Opcao = input("Insira a op√ß√£o: ")

        if Opcao == "1":
            print("\nVoc√™ escolheu a op√ß√£o: Street\n")
            Sexo()
        elif Opcao == "2":
            print("\nVoc√™ escolheu a op√ß√£o: Park\n")
            Sexo()
        elif Opcao == "0":
            print("Saindo do sistema...")
            break
        else:
            print("\nOp√ß√£o inv√°lida! Tente novamente.\n")

def Sexo():
    while True:
        print("\n         ===== <<< Sexo >>> =====")
        print("         |    [1] Feminino        |")
        print("         |    [2] Masculino       |")
        print("         |    [0] Voltar          |")
        print("           ------------------------------")
        Opcao = input("Insira a op√ß√£o: ")

        if Opcao in ["1", "2"]:
            if Opcao == "1":
                print("\nVoc√™ escolheu a op√ß√£o: Feminino\n")
            else:
                print("\nVoc√™ escolheu a op√ß√£o: Masculino\n")

            num_atletas = int(input("Quantos atletas participar√£o? "))
            num_juizes = 5
            juizes = ColetarNomesJuizes(num_juizes)
            ColetarNotas(num_atletas, juizes)
            break
        elif Opcao == "0":
            print("Voltando a sess√£o anterior...")
            break
        else:
            print("\nOp√ß√£o inv√°lida! Tente novamente.\n")

# Fun√ß√£o para coletar os nomes dos ju√≠zes
def ColetarNomesJuizes(num_juizes):
    juizes = []
    for j in range(1, num_juizes + 1):
        NomeJuiz = input(f"Digite o nome do juiz {j}: ").title()
        juizes.append(NomeJuiz)
    return juizes

# Definindo os crit√©rios de avalia√ß√£o com pesos iguais
CriteriosAvaliacao = {
    'Execu√ß√£o': 20,
    'Dificuldade da Manobra': 20,
    'Originalidade': 20,
    'Estilo': 20,
    'Uso da Rampa': 20
}

# Fun√ß√£o para coletar as notas de cada crit√©rio
def ColetarNotaCriterio(NomeJuiz, NomeAtleta, Criterio):
    Nota = float(input(f"\nNota do juiz {NomeJuiz} para {Criterio} do atleta {NomeAtleta} (0 a 20): "))
    while Nota < 0 or Nota > 20:
        Nota = float(input(f"Nota inv√°lida. Insira uma nota entre 0 e 20: "))
    return Nota

# Fun√ß√£o para calcular a nota final do juiz com base nos crit√©rios
def CalcularNotaJuiz(NotasCriterios):
    return sum(NotasCriterios.values())

# Fun√ß√£o para calcular a nota final removendo a maior e a menor nota
def CalcularNotaVolta(NotasJuizes):
    NotasJuizes.sort()  # Ordenar as notas
    NotasFiltradas = NotasJuizes[1:-1]  # Eliminar a menor e a maior nota
    return sum(NotasFiltradas)  # Somar as tr√™s notas restantes

# Fun√ß√£o para coletar notas para cada volta
def ColetarNotas(num_atletas, juizes):
    Voltas = {}
    resultados = []  # Lista para armazenar os resultados dos atletas

    for i in range(1, num_atletas + 1):  # Para cada atleta
        print(f"\nDigite as informa√ß√µes do Atleta {i}:")
        NomeAtleta = input("Digite o nome do atleta: ").title()
        PaisAtleta = input(f"Digite o pa√≠s do atleta {NomeAtleta}: ").title()

        NotasVolta = []  # Armazenar notas de todas as voltas para o atleta

        for j in range(1, 4):  # Para cada uma das 3 voltas
            print(f"\nInsira as notas dos ju√≠zes para a Volta {j}:")
            for NomeJuiz in juizes:  # Usar a lista de ju√≠zes coletada
                NotasCriterios = {}

                # Coletar as notas para cada crit√©rio
                for Criterio in CriteriosAvaliacao.keys():
                    Nota = ColetarNotaCriterio(NomeJuiz, NomeAtleta, Criterio)
                    NotasCriterios[Criterio] = Nota

                # Calcular a nota final do juiz para a volta
                NotaFinalJuiz = CalcularNotaJuiz(NotasCriterios)
                NotasVolta.append(NotaFinalJuiz)
                print(f"\nA nota final do atleta {NomeAtleta} ({PaisAtleta}) dada pelo juiz {NomeJuiz} √©: {NotaFinalJuiz:.2f}")

        # Armazenar as notas da volta
        Voltas[f"Atleta {i}"] = NotasVolta
        total_pontuacao = sum(NotasVolta)
        resultados.append({"nome": NomeAtleta, "pais": PaisAtleta, "pontuacao": total_pontuacao})

    # Exibir o p√≥dio
    ExibirPodio(resultados)

# Fun√ß√£o para exibir o p√≥dio dos melhores atletas
def ExibirPodio(resultados):
    # Ordenar os resultados pela pontua√ß√£o em ordem decrescente
    resultados_ordenados = sorted(resultados, key=lambda x: x['pontuacao'], reverse=True)

    # Exibe o p√≥dio
    print("\n    P√≥dio dos 3 Melhores Atletas:")
    if len(resultados_ordenados) > 0:
        print("\n            ü•á 1¬∫ Lugar")
        print(f"           {resultados_ordenados[0]['nome']} - {resultados_ordenados[0]['pontuacao']:.2f} ({resultados_ordenados[0]['pais']})")
    if len(resultados_ordenados) > 1:
        print("      ü•à 2¬∫ Lugar")
        print(f"    {resultados_ordenados[1]['nome']} - {resultados_ordenados[1]['pontuacao']:.2f} ({resultados_ordenados[1]['pais']})")
    if len(resultados_ordenados) > 2:
        print("      ü•â 3¬∫ Lugar")
        print(f"    {resultados_ordenados[2]['nome']} - {resultados_ordenados[2]['pontuacao']:.2f} ({resultados_ordenados[2]['pais']})")

    # Exibir a tabela com todos os resultados
    print("\nResultados Gerais:")
    print(f"{'Rank':<5} {'Nome':<20} {'Pa√≠s':<15} {'Pontua√ß√£o':<10}")
    for rank, atleta in enumerate(resultados_ordenados, start=1):
        print(f"{rank:<5} {atleta['nome']:<20} {atleta['pais']:<15} {atleta['pontuacao']:<10.2f}")
```
