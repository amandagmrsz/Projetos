```python
print("\033[1mPreencha seu currículo\033[m")
print()

# Solicita entrada dos dados pessoais
Nome = input("Digite seu nome: ")
Cpf = input("Digite seu CPF: ")
Endereco = input("Digite seu endereço: ")
EstadoCivil = input("Digite seu estado civil: ")
Telefone = input("Digite seu telefone: ")
Email = input("Digite seu email: ")
print()
Resumo = input("Sobre você (ex: sua experiência, habilidades e objetivos profissionais): ")

# Listas vazias
FormacaoAcademica = []
Experiencias = []
Cursos = []
Idiomas = []
print()

# Loop para entrada de dados da formação acadêmica
while True:
    Formacao = input("Digite sua formação  (formato: Nivel de Escolaridade - Curso - Instituição de Ensino) ou 'Q' para sair: ")
    if Formacao.upper() == "Q":
        break
    FormacaoAcademica.append(Formacao)
print()

# Exemplo para experiências:
while True:
    Experiencia = input("Digite sua experiência (formato: Cargo - Empresa - Tempo) ou 'Q' para sair: ")
    if Experiencia.upper() == "Q":
        break
    Experiencias.append(Experiencia)
print()

# Exemplo para cursos:
while True:
    Curso = input("Digite um curso que você fez (ou 'Q' para sair): ")
    if Curso.upper() == "Q":
        break
    Cursos.append(Curso)
print()

# Exemplo para idiomas:
while True:
    Idioma = input("Digite um idioma que você fala  (formato: Idioma - Nível) ou 'Q' para sair: ")
    if Idioma.upper() == "Q":
        break
    Idiomas.append(Idioma)
print()

print("\033[1mCurrículo\033[m")
print("==" *35)

# Exibindo os dados pessoais formatados
print(f"\033[1mNome:\033[m {Nome.title()}")
print(f"\033[1mCPF:\033[m {Cpf}")
print(f"\033[1mEndereço:\033[m {Endereco.upper()}")
print(f"\033[1mEstado Civil:\033[m {EstadoCivil.title()}")
print(f"\033[1mTelefone:\033[m {Telefone}")
print(f"\033[1mEmail:\033[m {Email}")
print("==" *35)
print()
print("\033[1mHabilidades & Competências\033[m")
print(Resumo)

print()
# Exibindo as listas formatadas
print("\033[1mFormação Acadêmica\033[m")
for item in FormacaoAcademica:
    print(f"• {item.title()}")
print()

print("\033[1mExperiências\033[m")
for item in Experiencias:
    print(f"• {item.title()}")
print()

print("\033[1mCursos\033[m")
for item in Cursos:
    print(f"• {item.title()}")
print()

print("\033[1mIdiomas\033[m")
for item in Idiomas:
    print(f"• {item.title()}")
```
